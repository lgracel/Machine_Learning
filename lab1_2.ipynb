```python
import matplotlib.pyplot as plt
import numpy as np

def func(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
x = np.arange(0, 16, .1)
y = func(x)
```

```python
M1 = np.array([[1, 1], [1, 15]])
v1 = np.array([func(1), func(15)])
w1 = np.linalg.solve(M1, v1)
y1 = w1[0] + w1[1]*x
```

```python
M2 = np.array([[1, 1, 1**2], [1, 8, 8**2], [1, 15, 15**2]])
v2 = np.array([func(1), func(8), func(15)])
w2 = np.linalg.solve(M2, v2)
y2 = w2[0] + w2[1]*x + w2[2]*(x**2)
```

```python
M3 = np.array([[1, 1, 1**2, 1**3], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
v3 = np.array([func(1), func(4), func(10), func(15)])
w3 = np.linalg.solve(M3, v3)
y3 = w2[0] + w3[1]*x + w3[2]*(x**2) + w3[3]*(x**3)
```

```python
print(np.around(w3, 2), '\n')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y, label = 'Initial')
plt.plot(x, y1, label = 'Linear')
plt.plot(x, y2, label = 'Quadratic')
plt.plot(x, y3, label = 'Cubic')
plt.legend()
plt.show()
```
