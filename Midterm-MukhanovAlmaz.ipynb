```python
#4 (b)
def func(img_array, ratio):
    for row_i, row_el in enumerate(img_array):
        for col_i, col_el in enumerate(row_el):
            img_array[row_i, col_i] = (col_el[0] * ratio[0] + col_el[1] * ratio[1] + col_el[2] * ratio[2])
    return img_array
img = np.array(misc.imread('pic.png'))
ratio = np.array([0.299, 0.587, 0.114])
#func(img, ratio)
misc.imsave('gray2_img.png', func(img,ratio))
```

```python
#5 (b)
def func(arr):
    digit_counter = {}
    for digit in arr:
        if digit_counter.get(digit) != None: #возвращает значение ключа, но если его нет, не бросает исключение, а возвращает default (по умолчанию None).
            digit_counter[digit] = digit_counter[digit] + 1
        else:
            digit_counter[digit] = 1

    keys = []
    values = []
    for key, value in digit_counter.items(): #возвращает пары (ключ, значение).
        keys.append(key)
        values.append(value)

    return (keys, values)
x = [2, 2, 2, 3, 3, 3, 5]
func(x)
```
